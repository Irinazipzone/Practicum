## –û–±—É—á–µ–Ω–∏–µ

–ü—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Ç–∞–∫—Å–∏ —è–≤–ª—è–µ—Ç—Å—è –∑–∞–¥–∞—á–µ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –î–ª—è –µ–µ —Ä–µ—à–µ–Ω–∏—è –æ–±—É—á–∏–º —Ç—Ä–∏ –º–æ–¥–µ–ª–∏ –ø–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞–º:
- –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å;  
- –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è;  
- –ì—Ä–∞–¥–∏–µ–Ω—Ç—ã–π –±—É—Å—Ç–∏–Ω–≥.  

–î–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç—Ä–∏–∫—É RMSE.

### –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

–î–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏: –∫–∞–ª–µ–Ω–¥–∞—Ä–Ω—ã–µ, ¬´–æ—Ç—Å—Ç–∞—é—â–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è¬ª –∏ —Å–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ, –î–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é make_features:

```python
def make_features(df, max_lag, rolling_mean_size):
    
    data = df.copy()
    
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek.astype('category')
    
    data.loc[(data.index.hour >= 0) & (data.index.hour < 6), 'times_of_day'] = 'night'
    data.loc[(data.index.hour >= 6) & (data.index.hour < 12), 'times_of_day'] = 'morning'
    data.loc[(data.index.hour >= 12) & (data.index.hour < 18), 'times_of_day'] = 'day'
    data.loc[(data.index.hour >= 18) & (data.index.hour < 24), 'times_of_day'] = 'evening'
    data['times_of_day'] = data['times_of_day'].astype('category')
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    data['lag_week'] = data['num_orders'].shift(24 * 7)
        
    data['rolling_mean_{}'.format(rolling_mean_size)] = data['num_orders'].shift().\
                                                    rolling(rolling_mean_size).mean()
    return data
```

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–î–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏ —Å–æ–±–µ—Ä–µ–º pipeline –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –æ–±—É—á–µ–Ω–∏—è.  
–î–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ —Ä–∞–∑–±–∏–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫, –∏—Å–ø–æ–ª—å–∑—É–µ–º –∫–ª–∞—Å—Å TimeSeriesSplit:

```python
tscv = TimeSeriesSplit(n_splits=10)
```

–î–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–±–Ω—É—é —Ç–∞–±–ª–∏—Ü—É total_ml:

```python
total_ml = pd.DataFrame(columns=['Name_ml', 'RMSE'])
```

#### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

```python
transformer_rfr = make_column_transformer(
    (
        OrdinalEncoder(
            dtype='int16',
            handle_unknown='use_encoded_value',
            unknown_value=-1
        ), make_column_selector(dtype_include=['category'])
    ),
    (
        StandardScaler(
        ), make_column_selector(dtype_include=['int16', 'float64'])
    ),
    remainder='passthrough'
)
```

```python
pipeline_rfr = make_pipeline(transformer_rfr, RandomForestRegressor(random_state=RANDOM_STATE))
```

```python
param_rfr = {
    'randomforestregressor__max_depth':np.arange(5, 21, 5),
    'randomforestregressor__min_samples_leaf': np.arange(2, 10)
}

reg_rfr = GridSearchCV(pipeline_rfr, param_rfr, scoring='neg_root_mean_squared_error', cv=tscv)

reg_rfr.fit(X_train, y_train) 
```

 <div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:  </b>
    
–û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è TimeSeriesSplit, –∞ –Ω–µ –ø—Ä–æ—Å—Ç–æ–π —Å–ø–ª–∏—Ç—Ç–µ—Ä

```python
total_ml = total_ml.append({'Name_ml': 'RandomForestRegressor', 'RMSE': np.sqrt((reg_rfr.best_score_)*(-1))},\
                           ignore_index=True)
total_ml
```

#### –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

–î–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä—ã OneHotEncoder –∏ StandardScaler:

```python
transformer_ln = make_column_transformer(
    (
        OneHotEncoder(
            dtype='uint8',
            handle_unknown='ignore'
        ), make_column_selector(dtype_include=['category'])
    ),
    (
        StandardScaler(
        ), make_column_selector(dtype_include=['int16', 'float64'])
    ),
    remainder='passthrough'
)
```

```python
pipeline_ln = make_pipeline(transformer_ln, LinearRegression())
```

```python
param_ln = {
    'linearregression__normalize':[False, True]
}

reg_ln = GridSearchCV(pipeline_ln, param_ln, scoring='neg_root_mean_squared_error', cv=tscv)

reg_ln.fit(X_train, y_train) 
```

```python
total_ml = total_ml.append({'Name_ml': 'LinearRegression', 'RMSE': np.sqrt((reg_ln.best_score_)*(-1))},\
                           ignore_index=True)
total_ml
```

#### –ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π –±—É—Å—Ç–∏–Ω–≥ Catboost

```python
pipeline_ctb = make_pipeline(transformer_rfr, CatBoostRegressor(loss_function='RMSE', silent=True,\
                                                         learning_rate=0.1, n_estimators=100,\
                                                         random_state=RANDOM_STATE))
```

```python
param_ctb = {'catboostregressor__max_depth': np.arange(8, 15)}
```

```python
reg_ctb = GridSearchCV(pipeline_ctb, param_ctb, scoring='neg_root_mean_squared_error', cv=tscv)
reg_ctb.fit(X_train,y_train)
```
