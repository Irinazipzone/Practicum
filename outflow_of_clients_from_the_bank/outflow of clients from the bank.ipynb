А вообще способов заполнения много, а какой из них выбрать — однозначного ответа нет. Но можно придерживаться примерно такого алгоритма:
* Если есть возможность, то, конечно, лучше выяснить природу этих пропусков, в реальных данных чаще всего это можно выяснить либо у коллег, либо на ресурсах, откуда ты берешь эти данные, либо просто по контексту и знанию предметной области.
* Если же такой возможности нет, то можно попробовать поискать корреляцию с другими признаками, сгруппировать данные по этим признакам и заполнить медианой/средним по группам.
* Если и тут все глухо, то можно заполнить и медианой/средним по всему признаку, как в этом случае, если есть уверенность, что распределение не исказится, для этого мы сравниваем характеристики распределения(mean, std, median) до и после заполнения
* На крайний случай, если пропусков не очень много, можно заполнить пропуски случайными значениями, задав вероятности равными частоте в исходном признаке. Тут будет полезен <a href="https://www.w3schools.com/python/ref_random_choices.asp" target="blank_">random.choices</a>. В параметр `weights` можно передать результат `value_counts(normalize=True)` для этого признака. А для самого заполнения можно использовать известный `fillna`.
* Иногда пропуски можно заполнить с помощью предсказаний моделей, сделав признак с пропусками целевым, либо, например, с помощью <a href="https://scikit-learn.org/stable/modules/generated/sklearn.impute.KNNImputer.html">KNNImputer</a>
* Но если адекватного способа заполнения, который не противоречит логике данных и здравому смыслу, не находится, то остается только удалить эти пропуски, либо даже удалить весь признак целиком. Для того, чтобы узнать, как именно тот или иной признак влияет на обучение модели, можно посмотреть на **feature_importances**, то есть важность признаков для решения модели. Если важность очень мала, то при удалении метрики поменяются незначительно, а значит мы сможем оставить драгоценные данные с остальными полезными признаками
    
При любом варианте заполнения, главное не забывать, что данные должны оставаться адекватными, приближенными к исходным значениям, а распределение не должно исказиться

### Пробное обучение

Проведем пробное обучение моделей без учета дисбаланса. Ограничимся двумя алгоритмами обучения - решающее дерево и случайный лес. Линейные алгоритмы не используем, так как линейные зависимоти признаков отсутствуют. В качестве основных метрик качества моделей выберем F1-меру и Auc-Roc. Предварительно создадим функцию расчета метрик качества.

Просмотрим линейные зависимости признаков:

```python
sns.pairplot(df, hue='exited')
```

Как видно из графиков, линейные зависимости отсутствуют. Но присутствуют бимодальные распределения признаков, что является благоприятным условием для обучения деревьев.

Создадим функцию рассчета метрик качества. Функция возвращает два значения: первое - значение F1-меры, второе - площадь под Roc-кривой:

```python
def score_func(name_model):
    predicted_valid = name_model.predict(features_valid)
    f1 = f1_score(target_valid, predicted_valid)
   
    probabilities_valid = name_model.predict_proba(features_valid)
    probabilities_valid = probabilities_valid[:, 1]
    auc_roc = roc_auc_score(target_valid, probabilities_valid)
    
    return f1, auc_roc
```

Создадим сводную таблицу результатов обучения моделей:

```python
scores = pd.DataFrame(columns=['Name_model', 'Name_balancing', 'F1', 'Roc-Auc'])
```

```python
model_rfc = RandomForestClassifier(random_state=RANDOM_STATE, criterion='entropy', max_depth=12, min_samples_leaf=2)
model_rfc.fit(features_train, target_train)

scores = scores.append({'Name_model': 'model_rfc', 'Name_balancing': '-', 'F1': score_func(model_rfc)[0],\
                        'Roc-Auc': score_func(model_rfc)[1]}, ignore_index=True)

score_func(model_rfc)
```
